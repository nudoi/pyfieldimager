```python
import pyfieldimager as pfi

pfi.__version__
```

Load images.

```python
orthophoto = '../20231018-orthophoto.tif'
dsm = '../20231018-dsm.tif'

# Create a new FieldImager object
fi = pfi.FieldImage(orthophoto=orthophoto, dsm=dsm, gsd=1.0)
```

```python
fi.show()
```

Cutting images. The cut method allows you to cut the image to a specific area. The area is defined by the coordinates of the upper left and lower right corners of the area.

```python
fi.crop_field(rotation=52, x_range=[1500, 3000], y_range=[2050, 3000])
```

```python
fi.show()
```

Select vegetation area by VARI index.

```python
# fi.select_index()
```

```python
fi.calc_index('VARI', threshold=0, reverse=False)
```

```python
fi.update()
```

```python
fi.show()
```

```python
fi.show_dsm()
```

```python
fi.show_full()
```

```python
fi.show_dsm_full()
```

Split field by grid.

```python
# fi.select_grid()
```

```python
fi.crop_grid(rotation=0, x_range=[0, 1500], y_range=[0, 940], x_split=10, y_split=15, grid_num=True)
```

```python
fi.show()
```

```python
fi.show_full()
```

```python
fi.show_dsm()
```

```python
field = fi.split()
```

```python
field[0][4].show()
```

```python
field[0][4].show_dsm()
```

Plant projective area

```python
field[0][4].proj_area()
```

Plant surface area

```python
field[0][4].surf_area()
```

Area of square

```python
field[0][4].square()
```
