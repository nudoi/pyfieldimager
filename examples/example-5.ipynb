```python
import pyfieldimager as pfi

pfi.__version__
```

```python
orthophoto = '../All-Purpose-Trail-2016-6-29-orthophoto.tif'
dsm        = '../All-Purpose-Trail-2016-6-29-dsm.tif'

# Create a new FieldImager object
fi = pfi.FieldImage(orthophoto=orthophoto, dsm=dsm, gsd=0.0669)
```

```python
fi.show()
```

```python
fi.show_dsm()
```
