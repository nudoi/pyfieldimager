```python
import pyfieldimager as pfi

pfi.__version__
```

Set orthophoto & DSM file path.

```python
# Orthophoto and dsm file paths
orthophoto = '../20231018-orthophoto.tif'
dsm = '../20231018-dsm.tif'

# Create a new FieldImage object
fi = pfi.FieldImage(orthophoto=orthophoto, dsm=dsm)
```

```python
# show orthophoto
fi.show()
```

```python
# show dsm
fi.show_dsm()
```

Select the field area. You can rotate field image.

```python
#fi.select_field() # GUI
fi.crop_field(rotation=52, x_range=[1350, 3650], y_range=[1750, 3480])
```

After select the area, update() to confirm.

```python
fi.update()
```

```python
fi.show()
```

```python
fi.show_dsm()
```

Show each RGB band and HSV model.

```python
fi.show_rgb()
#fi.show_rgb(band="R") # Red band
```

```python
fi.show_hue()
```

```python
fi.show_saturation()
```

```python
fi.show_value() # show_brightness() also available
```

Select vegetation index.

(about index, cf. https://www.opendronemap.org/fieldimager/#p6)

```python
fi.show()
```

Create DTM(Digital Terrain Model) from Soil DSM.

Interpolate missing value of Soil DSM by 3D spline. Methods are the same as in Pandas interpolate().

```python
fi.create_dtm(axis=1, method='spline', order=3)
```

```python
fi.show_dtm()
```
